```python
# Create Classification version of target variable
df['goodquality'] = [1 if x >= 7 else 0 for x in df['quality']]
X = df.drop(['quality','goodquality'], axis = 1)
y = df['goodquality']

df = df.drop('total sulfur dioxide', axis=1)
df.replace({'white': 1, 'red': 0}, inplace=True)
```

```python
df.head(17)
```

# Spliting The dataset :

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression()

model1.fit(X_train,y_train)

print('Training Accuracy : ', metrics.roc_auc_score(y_train, model1.predict(X_train)))

print('Validation Accuracy : ', metrics.roc_auc_score(y_test, model1.predict(X_test)))
```

```python
prediction1 = model1.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix

print(accuracy_score(prediction1,y_test))
```

```python
print(metrics.classification_report(y_test,model1.predict(X_test)))
```
